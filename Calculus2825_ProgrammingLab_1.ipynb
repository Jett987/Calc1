## 3 Secant Lines

Problem 3.)

```python
import math
import numpy as np

def g(x):
  return np.sin(2*2*math.pi/x)

g(1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(.1,2,10000)

plt.plot(x,g(x))
```

```python
Q = [2,1.5,1.1,1.01,1.001]

plt.plot(x,g(x))
for i in Q:
  plt.plot([1,i],[0,g(i)])
  plt.plot(i,g(i),'o')
```

Part a.)

Initially I do not think the slopes are approaching a limit. The values are too far away from the initial x value. 

```python
mylist = [2,1.5,1.1,1.01,1.001]

for i in mylist:
  print(g(i))
```

```python
mylist = [2,1.5,1.1,1.01,1.001]

for i in mylist:
  print((g(i)-g(1))/(i-1))
```

Part B.)

```python
Q = [2,1.5,1.1,1.01,1.001]

plt.plot(x,g(x))
for i in Q:
  plt.plot([1,i],[0,g(i)])
  plt.plot(i,g(i),'o')
```
